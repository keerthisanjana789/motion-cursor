**Linear Regression**


*   Single variable
*   Example: House Price Prediction



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#loading the data into the pandas dataframe
df = pd.read_csv("homeprices.csv")
df
```

```python
plt.scatter(df.area,df.price)
plt.xlabel('area')
plt.ylabel('price')
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
#fitting the data means training the regression model
#if the output doesn't show any error, that means this linear regression model is ready to predict the outputs
```

```python
plt.scatter(df.area,df.price)
plt.xlabel('area')
plt.ylabel('price')
plt.plot(df.area,model.predict(df[['area']]))
```

```python
#predicting the price of home whose area is 3300 square ft
model.predict([[3300]])
```

```python
#value of coefficient which is m
model.coef_
```

```python
#value of the intercept
model.intercept_
```

```python
#to get to know how the model pridicted the price of the house with area of 3300 as 628715.75342466 is because
#of the equation y = mx + b
#we have the values of m,x and b (here x is the area for which we are predicting the price)
135.78767123 * 3300 + 180616.43835616432
```
